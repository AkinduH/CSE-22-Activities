## Application 1 Part C

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 500, 1000)
y = 2.01 * np.sin(4.95 * x) - 1.9899 * np.sin(5 * x)

max_y = np.max(y)
max_x = x[np.argmax(y)]

plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Plot of 2.01sin(4.95x) - 1.9899sin(5x)')
plt.xlabel('x')
plt.ylabel('y')

plt.plot(max_x, max_y, 'ro', label=f'Max: ({max_x:.2f}, {max_y:.2f})')
plt.legend()
plt.grid(True)
plt.show()

print(f"The maximum value is approximately {max_y:.4f} at x  {max_x:.4f}")
```

```python
x = np.linspace(0, 1, 1000)[1:]
y = 2.01 * np.sin(4.95 * x) - 1.9899 * np.sin(5 * x)

break_points = np.where(np.diff(np.sign(y)))[0]

plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Plot of 2.01sin(4.95x) - 1.9899sin(5x)')
plt.xlabel('x')
plt.ylabel('y')

for bp in break_points:
    plt.axvline(x=x[bp], color='g', linestyle='--', label='Foundation break' if bp == break_points[0] else '')

plt.legend()
plt.grid(True)
plt.show()

print(f"The elastic foundation breaks at x  {', '.join([f'{x[bp]:.4f}' for bp in break_points])}")
```

## Application 3

```python
def E(h):
    return (1/(2 * 9.82)) * ((5 * 4 * 3**(1/4)) / (8 * h**(5/4)))**2 + h

h = np.linspace(0.1, 5, 1000)

E_values = E(h)

min_h = h[np.argmin(E_values)]
min_E = np.min(E_values)

plt.figure(figsize=(10, 6))
plt.plot(h, E_values)
plt.plot(min_h, min_E, 'ro', label=f'Min: ({min_h:.2f}, {min_E:.2f})')
plt.title('Plot of E vs h')
plt.xlabel('h')
plt.ylabel('E')
plt.ylim(0, 15)
plt.grid(True)
plt.legend()
plt.show()

print(f"The minimum value of E is approximately {min_E:.4f} at h  {min_h:.4f}")
```
